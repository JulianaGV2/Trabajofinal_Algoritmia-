```python
#Librerias necesarias para la codificación y el sistema operativo.

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#FUNCIONES

#La función mostrar, muestra la información en el dataframe de manera tabular.
"""La información corresponde a un diccionario llamado 'colmenas' que contiene 5 colmenas
con sus respectivas cantidad de abejas."""

def mostrar(colmenas):
  dfColmenas = pd.DataFrame({
        "colmenas" : colmenas.keys(),
        "abejas": colmenas.values()
    })
  print("--Estado de las colmenas en el Valle Floral--")
  display(dfColmenas)

#La función fumigar nos indica la colmena específica a fumigar.
"""Fumigar colmenas lleva a que la cantidad de abejas en cada colmenas se reduzca
en una cantidad de 30 abejas; pero si esta cantidad se reduce de tal manera que queda
en negativo, entonces lo dejamos en un valor igual a cero."""

def fumigar(colmenas):
  colmenaAFumigar = input("Ingrese el número de la colmena a fumigar (1,2,3...)):")
  clave = 'colmena'+ colmenaAFumigar
  abejas = colmenas[clave]
  fumigacion = abejas - 30
  if fumigacion < 0:
    fumigacion = 0
  colmenas[clave] = fumigacion
  return colmenas

#La función nos permite aplicar pesticidas a todas las colmenas.
"""Aplicar pesticidas significa que una colmena puede aumentar su cantidad en un 10%."""

def pesticidas(colmenas):
  for i in range(len(colmenas.values())):
    clave = 'colmena'+str(i+1)
    abejas = colmenas[clave]
    abejas = round(abejas + abejas*0.1)
    colmenas[clave] = abejas
  return colmenas

#La función rescata a una reina especial de las abejas.
"""Rescartar una reina especial significa quetodas las colmenas pueden aumentar
 sus cantidades en 50 abejas."""

def rescate_reina(colmenas):
  for i in range(len(colmenas.values())):
    clave = 'colmena'+str(i+1)
    abejas = colmenas[clave]
    abejas = abejas + 50
    colmenas[clave] = abejas
  return colmenas

#La función establece la migración de cierto número abejas entre dos colmenas.
def migracion(colmenas):
  colmena1 = input("Ingrese la colmena a la cual le va a extraer abejas (1,2,3...):")
  colmena2 = input("Ingrese la colmena a la cual le va a introducir abejas (1,2,3...):")
  num_abejas = int(input("Ingrese el número de abejas a movilizar:\n "))
  clave1 = 'colmena'+ colmena1
  clave2 = 'colmena'+ colmena2
  colmenas[clave1] = colmenas[clave1] - num_abejas
  colmenas[clave2] = colmenas[clave2] + num_abejas
  return colmenas

#La función nos muestra un diagrama de barras y un resumen de los datos que se muestran en el.
def estadisticas(colmenas):
  dfColmenas= pd.DataFrame({
      "colmenas" : colmenas.keys(),
      "abejas": colmenas.values()
    })
  analisis = dfColmenas.describe()
  print("A continuación observará la cantidad de colmenas, la media, desviación estandar, valor minimo, quartiles y valor máximo:")
  print(analisis)
  x = dfColmenas["colmenas"]
  y = dfColmenas["abejas"]
  plt.figure(figsize=(6,4))
  grafico = plt.bar(x,y, label='Estadisticas', alpha= 0.7, color='orange')
  plt.xticks(x,['1', '2', '3', '4', '5'])
  plt.xlabel('Colmenas')
  plt.ylabel('Abejas')
  plt.legend(loc=1)
  plt.savefig('Figura_Estadisticas', dpi=300, bbox_inches='tight')
  plt.show()

# Pregunta para la acción de libre elección en el proyecto
"""Si la cantidad de abejas en cada colmena representa un porcentaje de aporte para la permanencia de la
biodiversidad y reconstrucciòn del medio ambiente siempre y cuando estas sean mayor o igual a 200. ¿Cómo se observa
ésta distribución en función del total de la colmena?."""

#La función nos muestra un diagrama de torta que responde a la pregunta anterior.
def aporte(colmenas):
  abejas = list(colmenas.values())
  print(f"El total de abejas es:{sum(abejas)}")
  colmenas = list(colmenas.keys())
  colores = ['yellow','green','red','blue','pink']
  plt.figure()
  ganancia = plt.pie(abejas, labels= colmenas, colors=colores, autopct='%1.1f%%')
  plt.title('DISTRIBUCION DE APORTE AL PLANETA',fontsize=12)
  plt.axis('equal')
  plt.savefig('Figura_Nueva_Accion', dpi=300, bbox_inches='tight')
  print(f"El aporte respecto a la cantidad de abejas por colmena se distribuye de la siguiente manera:")
  plt.show()

# La función guarda la información final en el archivo de formato txt.
def guardar(colmenas):
  with open(ruta, 'w') as archivo_txt:
    for i in range(5):
      clave = 'colmena'+ str(i+1)
      archivo_txt.write("colmena")
      archivo_txt.write(str(i+1))
      archivo_txt.write(" ")
      archivo_txt.write(str(colmenas[clave]))
      archivo_txt.write('\n')

#CREACIÓN DE UN ARCHIVO DE TRABAJO

"""
Ejecutar y no modificar.
Creación de un archivo de texto plano para trabajar sobre este."""

directorio = "/content/"
archivo = "abejas.txt"
ruta = os.path.join(directorio, archivo)

with open(ruta, 'w') as archivo_txt:
  colmenas = []
  for i in range(5):
    abejas = input(f"Ingrese el número entero de abejas en la colmena {i+1}:")
    colmenas.append(abejas)
    archivo_txt.write("colmena")
    archivo_txt.write(str(i+1))
    archivo_txt.write(" ")
    archivo_txt.write(abejas)
    archivo_txt.write('\n')

colmenas = {}
with open(ruta, "r") as archivo_txt:
  for i in archivo_txt:
    (k, v) = i.split()
    colmenas[(k)] = int(v)
  print('Colmenas =', colmenas)

# Mientras booleano = True, el programa seguirá preguntando por una opción.
booleano = True
while booleano:
  mostrar(colmenas)
  print("--Opciones--")
  print("1) Fumigar una colmena")
  print("2) Pesticidas")
  print("3) Rescate a una reina especial de las abejas")
  print("4) Migración de algunas abejas entre colmenas")
  print("5) Mostrar estadisticas")

  action = int(input())
  if action == 1:
    fumigar(colmenas)

  if action == 2:
    colmenas = pesticidas(colmenas)

  if action == 3:
    colmenas = rescate_reina(colmenas)
    print(colmenas)

  if action == 4:
    colmenas = migracion(colmenas)

  if action == 5:
    print(estadisticas(colmenas))

  #Si el numero minimo de una colmena es mayor o igual a 200, indica que ya se gano el juego y se cierra el ciclo while
  if min(colmenas.values())>=200:
    mostrar(colmenas)
    booleano = False
    guardar(colmenas)
    print("¡Felicidades! Has salvado a las abejas del Valle Floral.")

    nuevAccion = input("¿Desea ver la distribucion del aporte al planeta por cantidad de abejas en colmenas? (responde si o no)\n ")
    if nuevAccion == "si":
      print(aporte(colmenas))
```
